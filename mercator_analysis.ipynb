```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
x = ["Australia", "India", "Greenland"]
y = [7633565, 2973190, 2166086]
plt.bar(x, y)
plt.xlabel("Country")
plt.ylabel("Land area (km^2)")
plt.title("Land area vs Country")
plt.savefig("greenlandcomp.png", bbox_inches="tight")
```

```python
plt.style.use('ggplot')
x = ["Russia", "Canada", "China", "United States"]
y = [16377742, 9093507, 9326410, 9147643]
plt.bar(x, y)
plt.xlabel("Country")
plt.ylabel("Land area (10 million km^2)")
plt.title("Land area vs Country")
plt.savefig("russiacomp.png", bbox_inches="tight")
```
